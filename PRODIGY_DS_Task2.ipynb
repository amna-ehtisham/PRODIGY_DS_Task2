# TASK 2: Data Cleaning and Exploratory Data Analysid on Titanic Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the Titanic dataset
titanic_df = pd.read_csv('/content/titanic.csv')

# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
# Check for missing values
print(titanic_df.isnull().sum())
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)


titanic_df.dropna(subset=['Fare'], inplace=True)

titanic_df['Sex'] = titanic_df['Sex'].astype('category')
titanic_df['Embarked'] = titanic_df['Embarked'].astype('category')

titanic_df.head()
```

```python
# Plot the survival rate
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Survived')
plt.title('Survival Count')
plt.show()
```

```python
# Plot the survival rate by gender
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Survived', hue='Sex')
plt.title('Survival Count by Gender')
plt.show()
```

```python
# Plot the survival rate by class
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Survived', hue='Pclass')
plt.title('Survival Count by Class')
plt.show()
```

```python
# Plot the survival rate by class
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Survived', hue='Pclass')
plt.title('Survival Count by Class')
plt.show()
```

```python
# Plot the age distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic_df, x='Age', kde=True, bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
# Plot the fare distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic_df, x='Fare', kde=True, bins=30)
plt.title('Fare Distribution')
plt.show()
```

```python
# Plot the survival rate by embarkation point
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic_df, x='Survived', hue='Embarked')
plt.title('Survival Count by Embarkation Point')
plt.show()
```
